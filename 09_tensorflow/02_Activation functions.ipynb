```python
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf
from tensorflow import keras

%matplotlib inline
```

```python
x = np.arange(-5, 5, 0.1)
```

```python
y = keras.activations.sigmoid(x)
plt.plot(x, y)
plt.show()
```

```python
def sigmoid(x):
    return 1 / (1 + tf.exp(-x))
```

```python
y = sigmoid(x)
plt.plot(x, y)
plt.grid()
plt.show()
```

```python
y = keras.activations.relu(x)
plt.plot(x, y)
plt.show()
```

```python
relu = lambda x: tf.maximum(0.0, x)
y = relu(x)
plt.plot(x, y)
plt.show()
```
